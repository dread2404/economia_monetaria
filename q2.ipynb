# 2. Ajustes

## 2.1 Deposito à Vista

```python
deposito_a_vista = deposito_a_vista.rename(
    columns={
        deposito_a_vista.columns[0]: "data",
        deposito_a_vista.columns[1]: "deposito_a_vista"
        }
    )

deposito_a_vista = deposito_a_vista.drop(deposito_a_vista.columns[-1], axis=1)
deposito_a_vista.head()
```

```python
# Mudando tipo de str para float
deposito_a_vista['deposito_a_vista'] = deposito_a_vista['deposito_a_vista'].str.replace(',','.').astype(float)
deposito_a_vista.head()
```

## 2.2 Meio de Pagamento

```python
meio_pagamento = meio_pagamento.rename(
    columns={
        meio_pagamento.columns[0]: "data",
        meio_pagamento.columns[1]: "meio_pagamento"
        }
    )

meio_pagamento = meio_pagamento.drop(meio_pagamento.columns[-1], axis=1)
meio_pagamento.head()
```

```python
# Mudando tipo de str para float
meio_pagamento['meio_pagamento'] = meio_pagamento['meio_pagamento'].str.replace(',','.').astype(float)
meio_pagamento.head()
```

## 2.3 Moeda em Poder Pública

```python
moeda_em_poder_publico = moeda_em_poder_publico.rename(
    columns={
        moeda_em_poder_publico.columns[0]: "data",
        moeda_em_poder_publico.columns[1]: "moeda_em_poder_publico"
        }
    )

moeda_em_poder_publico = moeda_em_poder_publico.drop(moeda_em_poder_publico.columns[-1], axis=1)
moeda_em_poder_publico.head()
```

```python
# Mudando tipo de str para float
moeda_em_poder_publico['moeda_em_poder_publico'] = moeda_em_poder_publico['moeda_em_poder_publico'].str.replace(',','.').astype(float)
moeda_em_poder_publico.head()
```

## 2.4 Moeda Emitida

```python
moeda_emitida = moeda_emitida.rename(
    columns={
        moeda_emitida.columns[0]: "data",
        moeda_emitida.columns[1]: "moeda_emitida"
        }
    )

moeda_emitida = moeda_emitida.drop(moeda_emitida.columns[-1], axis=1)
moeda_emitida.head()
```

```python
# Mudando tipo de str para float
moeda_emitida['moeda_emitida'] = moeda_emitida['moeda_emitida'].str.replace(',','.').astype(float)
moeda_emitida.head()
```

## 2.5 Reservar Bancária

```python
reservas_bancarias = reservas_bancarias.rename(
    columns={
        reservas_bancarias.columns[0]: "data",
        reservas_bancarias.columns[1]: "reservas_bancarias"
        }
    )

reservas_bancarias = reservas_bancarias.drop(reservas_bancarias.columns[-1], axis=1)
reservas_bancarias.head()
```

```python
reservas_bancarias['reservas_bancarias'] = reservas_bancarias['reservas_bancarias'].str.replace(',','.').astype(float)
reservas_bancarias.head()
```

# 2.6 Transações em Reais

```python
columns_to_adjust = transacoes_rs.columns.tolist()
columns_to_adjust.remove('Category')

for column in columns_to_adjust:
    transacoes_rs[column] = transacoes_rs[column].str.replace(',','.').astype(float)

transacoes_rs.head()
```

# 3. Cálculo - Multiplicador Monetário

## 3.1 Preferência por Papel-Moeda

```python
# preferencia por papel moeda = moeda em poder publico / meio de pagamento

pref_papel = moeda_em_poder_publico.merge(meio_pagamento, on='data', how='inner')
pref_papel['preferencia_por_papel_moeda'] = pref_papel['moeda_em_poder_publico'] / pref_papel['meio_pagamento']
pref_papel = pref_papel[["data", "preferencia_por_papel_moeda"]]
pref_papel.head()
```

## 3.2 Preferência por Depósito à Vista

```python
# preferência por depósito à vista = deposito a vista / meio de pagamento

pref_deposito = deposito_a_vista.merge(meio_pagamento, on='data', how='inner')
pref_deposito['preferencia_por_deposito_a_vista'] = pref_deposito['deposito_a_vista'] / pref_deposito['meio_pagamento']
pref_deposito = pref_deposito[["data", "preferencia_por_deposito_a_vista"]]
pref_deposito.head()
```

## 3.3 Taxa de Encaixe em Moeda Corrente

```python
# taxa de encaixe em moeda corrente = moeda emitida / deposito a vista

taxa_encaixe = moeda_emitida.merge(deposito_a_vista, on='data', how='inner')
taxa_encaixe['taxa_de_encaixe_em_moeda_corrente'] = taxa_encaixe['moeda_emitida'] / taxa_encaixe['deposito_a_vista']
taxa_encaixe = taxa_encaixe[["data", "taxa_de_encaixe_em_moeda_corrente"]]
taxa_encaixe.head()
```

## 3.4 Taxa de Reserva Bancária

```python
# taxa de reserva bancária = reservas bancárias / deposito a vista

taxa_reserva = reservas_bancarias.merge(deposito_a_vista, on='data', how='inner')
taxa_reserva['taxa_de_reserva_bancaria'] = taxa_reserva['reservas_bancarias'] / taxa_reserva['deposito_a_vista']
taxa_reserva = taxa_reserva[["data", "taxa_de_reserva_bancaria"]]
taxa_reserva.head()
```

# 3.5 Multiplicador Monetário

```python
multiplicador_monetario = pref_papel.merge(pref_deposito, on='data', how='inner')
multiplicador_monetario = multiplicador_monetario.merge(taxa_encaixe, on='data', how='inner')
multiplicador_monetario = multiplicador_monetario.merge(taxa_reserva, on='data', how='inner')
```

```python
multiplicador_monetario['multiplicador_monetario'] = 1 / (
    multiplicador_monetario["preferencia_por_papel_moeda"] 
    + 
    multiplicador_monetario["preferencia_por_deposito_a_vista"] 
    * (multiplicador_monetario["taxa_de_encaixe_em_moeda_corrente"] + multiplicador_monetario["taxa_de_reserva_bancaria"])
)

multiplicador_monetario.head()
```

```python
# Convertendo a coluna 'data' para datetime
# Garantindo que o formato da data seja consistente

multiplicador_monetario['data'] = multiplicador_monetario['data'].astype(str).str.replace('.', '-')
multiplicador_monetario['data'] = multiplicador_monetario['data'].apply(lambda x: x + '0' if len(x) < 7 else x)
multiplicador_monetario

multiplicador_monetario['data'] = pd.to_datetime(multiplicador_monetario['data'], format='%Y-%m')

sns.lineplot(x='data', y='multiplicador_monetario', data=multiplicador_monetario)

plt.title('Multiplicador Monetário ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Multiplicador Monetário')

plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.figure(figsize=(20,15))

plt.show()
```

# 4. Quantidade de Transações x Tipo x Tempo

```python
transacoes_long = pd.melt(transacoes_qtde, 
                          id_vars=['Category'], 
                          var_name='Tipo_Transacao', 
                          value_name='Quantidade')

plt.figure(figsize=(14, 8))
sns.lineplot(data=transacoes_long, x='Category', y='Quantidade', hue='Tipo_Transacao', marker='o')

plt.title('Quantidade de Transações por Categoria e Tipo')
plt.xlabel('Categoria (Trimestre)')
plt.ylabel('Quantidade de Transações')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Tipo de Transação', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
